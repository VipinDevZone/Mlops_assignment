```python
# Que 4
def exponent(base, exp):
    # Initialize result to 1
    result = 1
    # Multiply the base with itself 'exp' times
    for _ in range(exp):
        result *= base
    return result

# Example usage
base = 5
exp = 4
result = exponent(base, exp)
print(f"{base} raised to the power of {exp} is: {result}")
```

```python
# Que 5
# Step 1: Create and write content to 'test.txt'
with open('test.txt', 'w') as file:
    file.write("line1\n")
    file.write("line2\n")
    file.write("line3\n")
    file.write("line4\n")
    file.write("line5\n")
    file.write("line6\n")
    file.write("line7\n")

# Step 2: Read from 'test.txt', skip line 5, and write to 'new_file.txt'
def skip_line_and_write(source_file, target_file, line_to_skip):
    with open(source_file, 'r') as infile:
        lines = infile.readlines()

    with open(target_file, 'w') as outfile:
        for i, line in enumerate(lines, start=1):
            if i != line_to_skip:
                outfile.write(line)

# Specify line to skip (5 in this case)
line_to_skip = 5

# Call the function to process the files
skip_line_and_write('test.txt', 'new_file.txt', line_to_skip)

print("Content copied to 'new_file.txt' with line 5 skipped.")
```

```python
#Que 8
import re

def find_words_with_letters_and_digits(input_string):
    # Split the input string into words
    words = input_string.split()

    # Define a list to hold words containing both letters and digits
    result = []

    # Iterate through each word
    for word in words:
        # Check if the word contains both letters and digits
        if re.search(r'[A-Za-z]', word) and re.search(r'\d', word):
            result.append(word)

    return result

# Given input string
str1 = "Harry25 is Gryffindor Wizard50 and cursed"

# Find and print words that contain both letters and digits
words_with_letters_and_digits = find_words_with_letters_and_digits(str1)
for word in words_with_letters_and_digits:
    print(word)
```

```python
#Que 9
import re

def substitute_special_characters(input_string):
    # Replace each special character with '#'
    result = re.sub(r'[^a-zA-Z0-9\s]', '#', input_string)
    return result

# Given input string
str1 = '/*Modi is @Politician & Melony!!'

# Substitute special characters and print the result
output_string = substitute_special_characters(str1)
print(output_string)
```

```python
#Que 11
# Define two sets A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# Perform set operations
A_minus_B = A - B           # Elements in A but not in B
B_minus_A = B - A           # Elements in B but not in A
A_intersection_B = A & B    # Elements common to both A and B
A_union_B = A | B           # All elements in either A or B

# Print the results
print("A - B:", A_minus_B)
print("B - A:", B_minus_A)
print("A ∩ B:", A_intersection_B)
print("A ∪ B:", A_union_B)
```

```python
#Que 13
# Define the parent class Vehicle
class Vehicle:
    def __init__(self, max_speed, mileage):
        """
        Initialize a new Vehicle instance.

        Parameters:
        max_speed (int): The maximum speed of the vehicle.
        mileage (int): The mileage of the vehicle.
        """
        self.max_speed = max_speed
        self.mileage = mileage

    def display_info(self):
        """
        Display the max speed and mileage of the vehicle.
        """
        print(f"Max Speed: {self.max_speed} km/h")
        print(f"Mileage: {self.mileage} km")

# Define the child class Bus that inherits from Vehicle
class Bus(Vehicle):
    def __init__(self, max_speed, mileage, seating_capacity):
        """
        Initialize a new Bus instance. Inherits from Vehicle.

        Parameters:
        max_speed (int): The maximum speed of the bus.
        mileage (int): The mileage of the bus.
        seating_capacity (int): The seating capacity of the bus.
        """
        # Initialize the parent class (Vehicle) with max_speed and mileage
        super().__init__(max_speed, mileage)
        self.seating_capacity = seating_capacity

    def display_info(self):
        """
        Display the max speed, mileage, and seating capacity of the bus.
        """
        # Call the parent class method to display max speed and mileage
        super().display_info()
        print(f"Seating Capacity: {self.seating_capacity}")

# Example usage:
# Create a Bus instance
bus = Bus(max_speed=120, mileage=30000, seating_capacity=50)

# Display the bus's information
bus.display_info()



```

```python
#Que 14
# Define the parent class Vehicle
class Vehicle:
    def __init__(self, name, max_speed, mileage):
        """
        Initialize a new Vehicle instance.

        Parameters:
        name (str): The name of the vehicle.
        max_speed (int): The maximum speed of the vehicle.
        mileage (int): The mileage of the vehicle.
        """
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage

    def seating_capacity(self, capacity):
        """
        Return the seating capacity of the vehicle.

        Parameters:
        capacity (int): The seating capacity of the vehicle.

        Returns:
        str: A string describing the seating capacity.
        """
        return f"The seating capacity of a {self.name} is {capacity} passengers"

# Define the child class Bus that inherits from Vehicle
class Bus(Vehicle):
    def __init__(self, name, max_speed, mileage, seating_capacity=50):
        """
        Initialize a new Bus instance. Inherits from Vehicle.

        Parameters:
        name (str): The name of the bus.
        max_speed (int): The maximum speed of the bus.
        mileage (int): The mileage of the bus.
        seating_capacity (int, optional): The seating capacity of the bus. Defaults to 50.
        """
        # Initialize the parent class (Vehicle) with name, max_speed, and mileage
        super().__init__(name, max_speed, mileage)
        self.seating_capacity_value = seating_capacity

    def seating_capacity(self):
        """
        Return the seating capacity of the bus with the default value.

        Returns:
        str: A string describing the seating capacity of the bus.
        """
        return f"The seating capacity of a {self.name} is {self.seating_capacity_value} passengers"

# Example usage:
# Create a Bus instance
bus = Bus(name="bus", max_speed=120, mileage=30000)

# Display the seating capacity of the bus
print(bus.seating_capacity())
```

```python
#Que 15
class Vehicle:
    def __init__(self, name, mileage, capacity):
        """
        Initialize a new Vehicle instance.

        Parameters:
        name (str): The name of the vehicle.
        mileage (int): The mileage of the vehicle.
        capacity (int): The seating capacity of the vehicle.
        """
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        """
        Calculate the fare for the vehicle.

        Returns:
        float: The fare based on seating capacity.
        """
        return self.capacity * 100

class Bus(Vehicle):
    def __init__(self, name, mileage, capacity=50):
        """
        Initialize a new Bus instance. Inherits from Vehicle.

        Parameters:
        name (str): The name of the bus.
        mileage (int): The mileage of the bus.
        capacity (int, optional): The seating capacity of the bus. Defaults to 50.
        """
        super().__init__(name, mileage, capacity)

    def fare(self):
        """
        Calculate the fare for the bus including a 10% maintenance fee.

        Returns:
        float: The total fare with the additional maintenance fee.
        """
        total_fare = super().fare()
        maintenance_fee = total_fare * 0.10
        return total_fare + maintenance_fee

# Example usage:
# Create a Bus instance
school_bus = Bus(name="School Volvo", mileage=12)

# Display the fare for the bus
print("Total Bus fare is:", school_bus.fare())
```
